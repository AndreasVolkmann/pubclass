```python
import pandas as pd
%matplotlib inline
data = pd.read_csv("data.csv", quotechar='"')
```

```python
X = data.Text
y = data.Class
print(X.shape)
print(y.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words="english")
vect.fit(X)
train_dtm = vect.transform(X)
train_dtm
```

```python
pd.DataFrame(train_dtm.toarray(), columns=vect.get_feature_names())
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
vect = CountVectorizer(stop_words="english")
X_train_dtm = vect.fit_transform(X_train)
X_train_dtm
```

```python
X_test_dtm = vect.transform(X_test)
X_test_dtm
```

```python
# Build and evaluate
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
%time nb.fit(X_train_dtm, y_train)
```

```python
y_pred_class = nb.predict(X_test_dtm)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
labels = ["Endocrinology", "Gynocology", "Dermatology"]
cm = metrics.confusion_matrix(y_test, y_pred_class, labels=labels)
print(cm)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title("Confusion Matrix")
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
%time lg.fit(X_train_dtm, y_train)
```

```python
y_pred_class = lg.predict(X_test_dtm)
metrics.accuracy_score(y_test, y_pred_class)
```
